```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
import pickle
```

```python
ds = pd.read_csv('bdata_dummies.csv')
```

```python
ds
```

```python
columns = ds.columns.drop(labels=['book_title', 'book_authors', 'book_desc', 'book_title', 'genres', 'emotion_index'])
```

```python
X = ds[[c for c in columns]]
y = ds['book_rating']
```

```python
# check-up
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

X_train = X[0:3325]
X_test = X[3325:]
y_train = y[0:3325]
y_test = y[3325:]
md = Ridge(alpha=0.1).fit(X_train, y_train)
pred_rd = md.predict(X_test)
print('ridge')
print(np.sqrt(mean_squared_error(y_test, pred_rd)))
print(r2_score(y_test, pred_rd))
```

```python
model = Ridge(alpha=0.1).fit(X, y)
```

```python
pickle.dump(model, open('model.pkl','wb'))
```
