```python
indices = [i for i, x in enumerate(emotions) if x == {'Angry': 0, 'Fear': 0, 'Happy': 0, 'Sad': 0, 'Surprise': 0}]
```

```python
indices
```

```python
angry_list = list()
fear_list = list()
happy_list = list()
sad_list = list()
surprise_list = list()

for i in range(len(emotions)):
    angry_list.append(emotions[i]['Angry'])
    fear_list.append(emotions[i]['Fear'])
    happy_list.append(emotions[i]['Happy'])
    sad_list.append(emotions[i]['Sad'])
    surprise_list.append(emotions[i]['Surprise'])
```

```python
ds['anger'] = angry_list
ds['fear'] = fear_list
ds['happiness'] = happy_list
ds['sadness'] = sad_list
ds['surprise'] = surprise_list
ds
```

```python
for i in indices:
  ds.drop(labels=i, axis=0, inplace=True)
ds
```

```python
ds.to_csv(r'result_book_data.csv', index = False, header=True)
```
